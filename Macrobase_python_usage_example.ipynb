# Usage

Data preprocess:

```python
import pandas as pd
df = pd.read_csv("model1.csv", header=0)
# Add an "Outlier" column. The outlier here is defined as true_value != predicted_value.
df["Outlier"] = df['true_value'] != df["predicted_value"]
# Drop true class and predicted class columns.
df = df.drop(columns=["true_value", "predicted_value"])
df
```

Import Macrobase_python and use it:

```python
import Macrobase_python
```

```python
# Input: a discretized dataframe with an outlier column, and without true class and predicted class.
# min_support is the minimum support, and min_rr is the minimum risk ratio.
macrobase = Macrobase_python.Macrobase(df, min_support=0.1, min_rr=1.5)
macrobase.fit()
```

The output is a pandas dataframe. Each row indicates a frequent itemset with its support, accuracy, and risk ratio.
